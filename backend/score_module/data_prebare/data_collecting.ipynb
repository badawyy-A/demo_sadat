```python
import pandas as pd 

speed1 = pd.read_csv('speed.csv')
speed2 = pd.read_csv('speed_0.csv')
speed2.columns = ['Age', 'L1 (Work \nHarder)', 'L2 (Must \nImprove)',
       'L3 (Can do \nBetter)', 'L4 (Good)', 'L5 (Very \nGood)',
       'L6 (Athletic)  L7 (Sports', 'Fit)']

speed = [speed1 , speed2]

speed_table = pd.concat(speed , ignore_index=True)

speed_table.to_csv('speed_table.csv')

speed_table
```

```python
for i in speed_table.columns[1:]:
    speed_table[f'{i}_min'] = speed_table[i].str.split(pat='to').str[1]
    speed_table[f'{i}_max'] = speed_table[i].str.split(pat='to').str[0]
    speed_table.drop(columns=[i] , inplace=True)

speed_table
```

```python
for i in speed_table.columns[1:]:
    speed_table[i] = speed_table[i].astype(str).str.replace("\n", " ", regex=True)
    speed_table[i] = speed_table[i].astype(str).str.replace("<", "")
    speed_table[i] = speed_table[i].astype(str).str.replace("m", " ")
    speed_table[i] = speed_table[i].astype(str).str.replace("s", " ")
    speed_table[i] = speed_table[i].astype(str).str.replace("ms", "")
    speed_table[i] = speed_table[i].astype(str).str.replace(r"\s+", ",", regex=True)
    speed_table[i] = speed_table[i].str.replace(r"^\s*,|,\s*$", "", regex=True)



speed_table
```
